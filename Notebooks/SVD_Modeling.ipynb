```python
from surprise.model_selection import cross_validate
from surprise.prediction_algorithms import SVD
from surprise.model_selection import GridSearchCV
```

```python
svd = SVD()
svd.fit(trainset)
baseline_prediction = svd.test(testset)
accuracy.rmse(baseline_prediction)
accuracy.mse(baseline_prediction)
```

#### SVD Model Tuning 

```python
svd_params = {'n_factors': [10, 20, 50, 80],
             'reg_all': [0.02,0.04, 0.06],
             'n_epochs': [5, 10, 20, 30],
             'lr_all': [.002, .005]}
svd_grid = GridSearchCV(SVD, param_grid=svd_params)
```

```python
svd_grid.fit(data)
```

```python
print(svd_grid.best_score)
print(svd_grid.best_params)
```

```python
tuned_svd = SVD(n_factors= 80, reg_all= 0.06, n_epochs=30, lr_all=0.005)
tuned_svd.fit(trainset)
tuned_pred= tuned_svd.test(testset)
accuracy.rmse(tuned_pred)
accuracy.mae(tuned_pred)
```

```python
tuned_crossval = cross_validate(tuned_svd, data, measures=['rmse','mae'],cv=5)
print(tuned_crossval['test_rmse'])
print(tuned_crossval['test_mae'])
print(tuned_crossval['fit_time'])
print(tuned_crossval['test_time'])
print('-----------------------')
print(np.mean(tuned_crossval['test_mae']))
```
